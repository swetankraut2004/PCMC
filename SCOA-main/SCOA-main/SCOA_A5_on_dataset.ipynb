```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.preprocessing import MinMaxScaler
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error, mean_absolute_error, r2_score
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, Dropout, LSTM
from tensorflow.keras.optimizers import Adam
from tensorflow.keras.callbacks import EarlyStopping
import warnings
warnings.filterwarnings('ignore')

df = pd.read_csv('SCOA_A5.csv')

print("="*80)
print("STOCK MARKET TREND PREDICTION USING ARTIFICIAL NEURAL NETWORK")
print("="*80)

print("\n1. DATA OVERVIEW")
print("-"*80)
print(f"Dataset Shape: {df.shape}")
print(f"\nColumns: {list(df.columns)}")
print("\nFirst 10 rows:")
print(df.head(10))
print("\nDataset Info:")
print(df.info())
print("\nStatistical Summary:")
print(df.describe())
print("\nMissing Values:")
print(df.isnull().sum())
print("\nTarget variable (close price) statistics:")
print(f"  Min: ${df['close'].min():.2f}")
print(f"  Max: ${df['close'].max():.2f}")
print(f"  Mean: ${df['close'].mean():.2f}")
print(f"  Std: ${df['close'].std():.2f}")

print("\n2. DATA PREPROCESSING")
print("-"*80)

df['date'] = pd.to_datetime(df['date'], errors='coerce')
df = df.sort_values('date').reset_index(drop=True)
df = df.dropna()

print(f"Shape after cleaning: {df.shape}")
print(f"Date range: {df['date'].min()} to {df['date'].max()}")

print("\n3. FEATURE ENGINEERING")
print("-"*80)

df['price_change'] = df['close'] - df['open']
df['price_change_pct'] = ((df['close'] - df['open']) / df['open']) * 100
df['high_low_diff'] = df['high'] - df['low']
df['daily_return'] = df['close'].pct_change() * 100

df['MA_5'] = df['close'].rolling(window=5).mean()
df['MA_10'] = df['close'].rolling(window=10).mean()
df['MA_20'] = df['close'].rolling(window=20).mean()

df['volatility'] = df['close'].rolling(window=5).std()

delta = df['close'].diff()
gain = (delta.where(delta > 0, 0)).rolling(window=14).mean()
loss = (-delta.where(delta < 0, 0)).rolling(window=14).mean()
rs = gain / loss
df['RSI'] = 100 - (100 / (1 + rs))

df['volume_change'] = df['volume'].pct_change() * 100

df['upper_shadow'] = df['high'] - df[['open', 'close']].max(axis=1)
df['lower_shadow'] = df[['open', 'close']].min(axis=1) - df['low']

df['target'] = df['close'].shift(-1)

df = df.dropna()

print(f"Features created: {len(df.columns)}")
print(f"Shape after feature engineering: {df.shape}")
print(f"\nNew features:")
feature_list = ['price_change', 'price_change_pct', 'high_low_diff', 'daily_return',
                'MA_5', 'MA_10', 'MA_20', 'volatility', 'RSI', 'volume_change',
                'upper_shadow', 'lower_shadow', 'target']
for feat in feature_list:
    print(f"  - {feat}")

print("\n4. FEATURE SELECTION")
print("-"*80)

feature_columns = ['open', 'high', 'low', 'close', 'volume',
                   'price_change', 'price_change_pct', 'high_low_diff',
                   'daily_return', 'MA_5', 'MA_10', 'MA_20',
                   'volatility', 'RSI', 'volume_change',
                   'upper_shadow', 'lower_shadow']

X = df[feature_columns].values
y = df['target'].values

print(f"Feature matrix shape: {X.shape}")
print(f"Target vector shape: {y.shape}")
print(f"\nFeatures used for prediction:")
for i, col in enumerate(feature_columns):
    print(f"  {i+1}. {col}")

print("\n5. DATA NORMALIZATION")
print("-"*80)

scaler_X = MinMaxScaler(feature_range=(0, 1))
scaler_y = MinMaxScaler(feature_range=(0, 1))

X_scaled = scaler_X.fit_transform(X)
y_scaled = scaler_y.fit_transform(y.reshape(-1, 1))

print("Features normalized using MinMaxScaler (0-1 range)")
print("Target variable normalized using MinMaxScaler (0-1 range)")
print(f"Original target range: ${y.min():.2f} - ${y.max():.2f}")
print(f"Scaled target range: {y_scaled.min():.4f} - {y_scaled.max():.4f}")
print(f"Target mean: Original=${y.mean():.2f}, Scaled={y_scaled.mean():.4f}")

print("\n6. TRAIN-TEST SPLIT")
print("-"*80)

X_train, X_test, y_train, y_test = train_test_split(
    X_scaled, y_scaled, test_size=0.2, random_state=42, shuffle=False
)

print(f"Training set size: {X_train.shape[0]} samples ({X_train.shape[0]/len(X_scaled)*100:.1f}%)")
print(f"Test set size: {X_test.shape[0]} samples ({X_test.shape[0]/len(X_scaled)*100:.1f}%)")
print(f"Feature dimensions: {X_train.shape[1]}")

print("\n7. BUILDING ARTIFICIAL NEURAL NETWORK")
print("-"*80)

model = Sequential()

model.add(Dense(256, activation='relu', input_dim=X_train.shape[1],
                kernel_initializer='he_normal'))
model.add(Dropout(0.3))

model.add(Dense(128, activation='relu', kernel_initializer='he_normal'))
model.add(Dropout(0.3))

model.add(Dense(64, activation='relu', kernel_initializer='he_normal'))
model.add(Dropout(0.2))

model.add(Dense(32, activation='relu', kernel_initializer='he_normal'))
model.add(Dropout(0.2))

model.add(Dense(16, activation='relu', kernel_initializer='he_normal'))

model.add(Dense(1, activation='sigmoid'))

model.compile(optimizer=Adam(learning_rate=0.0005),
              loss='mean_squared_error',
              metrics=['mean_absolute_error'])

print("Neural Network Architecture:")
model.summary()

print("\n8. TRAINING THE MODEL")
print("-"*80)

early_stopping = EarlyStopping(monitor='val_loss', patience=30,
                               restore_best_weights=True, verbose=1)

history = model.fit(
    X_train, y_train,
    epochs=5,
    batch_size=64,
    validation_split=0.2,
    callbacks=[early_stopping],
    verbose=1
)

print("\nTraining completed!")
print(f"Total epochs run: {len(history.history['loss'])}")
print(f"Final training loss: {history.history['loss'][-1]:.6f}")
print(f"Final validation loss: {history.history['val_loss'][-1]:.6f}")
print(f"Best validation loss: {min(history.history['val_loss']):.6f}")

if len(history.history['loss']) < 300:
    print(f"Early stopping triggered - no improvement for 30 epochs")

print("\n9. MODEL EVALUATION")
print("-"*80)

train_predictions = model.predict(X_train)
test_predictions = model.predict(X_test)

train_predictions_original = scaler_y.inverse_transform(train_predictions)
test_predictions_original = scaler_y.inverse_transform(test_predictions)
y_train_original = scaler_y.inverse_transform(y_train)
y_test_original = scaler_y.inverse_transform(y_test)

train_mse = mean_squared_error(y_train_original, train_predictions_original)
train_rmse = np.sqrt(train_mse)
train_mae = mean_absolute_error(y_train_original, train_predictions_original)
train_r2 = r2_score(y_train_original, train_predictions_original)

test_mse = mean_squared_error(y_test_original, test_predictions_original)
test_rmse = np.sqrt(test_mse)
test_mae = mean_absolute_error(y_test_original, test_predictions_original)
test_r2 = r2_score(y_test_original, test_predictions_original)

print("Training Set Performance:")
print(f"  MSE:  {train_mse:.6f}")
print(f"  RMSE: {train_rmse:.6f}")
print(f"  MAE:  {train_mae:.6f}")
print(f"  R² Score: {train_r2:.6f}")

print("\nTest Set Performance:")
print(f"  MSE:  {test_mse:.6f}")
print(f"  RMSE: {test_rmse:.6f}")
print(f"  MAE:  {test_mae:.6f}")
print(f"  R² Score: {test_r2:.6f}")

mape = np.mean(np.abs((y_test_original - test_predictions_original) / y_test_original)) * 100
print(f"\nMean Absolute Percentage Error (MAPE): {mape:.2f}%")

print("\n10. PREDICTION ACCURACY ANALYSIS")
print("-"*80)

accuracy_threshold = 0.02
within_threshold = np.abs(test_predictions_original - y_test_original) / y_test_original <= accuracy_threshold
accuracy_percentage = (np.sum(within_threshold) / len(y_test_original)) * 100

print(f"Predictions within {accuracy_threshold*100}% of actual: {accuracy_percentage:.2f}%")
print(f"Average prediction error: ${test_mae:.4f}")

print("\n11. TREND PREDICTION")
print("-"*80)

actual_trend = np.sign(y_test_original[1:] - y_test_original[:-1])
predicted_trend = np.sign(test_predictions_original[1:] - test_predictions_original[:-1])

trend_accuracy = np.mean(actual_trend == predicted_trend) * 100

print(f"Trend Direction Accuracy: {trend_accuracy:.2f}%")
print("(Ability to predict if stock will go UP or DOWN)")

up_predictions = np.sum(predicted_trend > 0)
down_predictions = np.sum(predicted_trend < 0)
neutral_predictions = np.sum(predicted_trend == 0)

print(f"\nPredicted Trends in Test Set:")
print(f"  Upward trends: {up_predictions} ({up_predictions/len(predicted_trend)*100:.1f}%)")
print(f"  Downward trends: {down_predictions} ({down_predictions/len(predicted_trend)*100:.1f}%)")
print(f"  Neutral: {neutral_predictions} ({neutral_predictions/len(predicted_trend)*100:.1f}%)")

print("\n12. SAMPLE PREDICTIONS")
print("-"*80)

sample_indices = np.random.choice(len(y_test_original), min(10, len(y_test_original)), replace=False)
sample_indices = np.sort(sample_indices)

print(f"{'Actual Price':<15} {'Predicted Price':<15} {'Error':<15} {'% Error':<15}")
print("-"*60)
for idx in sample_indices:
    actual = y_test_original[idx][0]
    predicted = test_predictions_original[idx][0]
    error = predicted - actual
    pct_error = (error / actual) * 100
    print(f"${actual:<14.2f} ${predicted:<14.2f} ${error:<14.2f} {pct_error:<14.2f}%")

print("\n13. FUTURE PREDICTION FUNCTION")
print("-"*80)

def predict_next_day(last_data, model, scaler_X, scaler_y):
    last_data_scaled = scaler_X.transform(last_data.reshape(1, -1))
    prediction_scaled = model.predict(last_data_scaled, verbose=0)
    prediction = scaler_y.inverse_transform(prediction_scaled)
    return prediction[0][0]

last_row_features = X[-1]
next_day_prediction = predict_next_day(last_row_features, model, scaler_X, scaler_y)

last_close_price = df['close'].iloc[-1]
predicted_change = next_day_prediction - last_close_price
predicted_change_pct = (predicted_change / last_close_price) * 100

print(f"Last known closing price: ${last_close_price:.2f}")
print(f"Next day predicted price: ${next_day_prediction:.2f}")
print(f"Expected change: ${predicted_change:.2f} ({predicted_change_pct:+.2f}%)")

if predicted_change > 0:
    trend = "UPWARD ↑"
    recommendation = "Consider BUYING"
elif predicted_change < 0:
    trend = "DOWNWARD ↓"
    recommendation = "Consider SELLING or HOLDING"
else:
    trend = "NEUTRAL →"
    recommendation = "HOLD position"

print(f"Predicted trend: {trend}")
print(f"Trading recommendation: {recommendation}")

print("\n14. VISUALIZATIONS")
print("-"*80)

fig = plt.figure(figsize=(16, 12))

plt.subplot(3, 2, 1)
plt.plot(history.history['loss'], label='Training Loss', linewidth=2)
plt.plot(history.history['val_loss'], label='Validation Loss', linewidth=2)
plt.title('Model Loss During Training', fontsize=12, fontweight='bold')
plt.xlabel('Epoch')
plt.ylabel('Loss (MSE)')
plt.legend()
plt.grid(True, alpha=0.3)

plt.subplot(3, 2, 2)
plt.plot(history.history['mean_absolute_error'], label='Training MAE', linewidth=2)
plt.plot(history.history['val_mean_absolute_error'], label='Validation MAE', linewidth=2)
plt.title('Mean Absolute Error During Training', fontsize=12, fontweight='bold')
plt.xlabel('Epoch')
plt.ylabel('MAE')
plt.legend()
plt.grid(True, alpha=0.3)

plt.subplot(3, 2, 3)
plt.plot(y_test_original, label='Actual Prices', linewidth=2, alpha=0.7)
plt.plot(test_predictions_original, label='Predicted Prices', linewidth=2, alpha=0.7)
plt.title('Actual vs Predicted Stock Prices (Test Set)', fontsize=12, fontweight='bold')
plt.xlabel('Sample Index')
plt.ylabel('Stock Price ($)')
plt.legend()
plt.grid(True, alpha=0.3)

plt.subplot(3, 2, 4)
plt.scatter(y_test_original, test_predictions_original, alpha=0.5)
plt.plot([y_test_original.min(), y_test_original.max()],
         [y_test_original.min(), y_test_original.max()],
         'r--', linewidth=2, label='Perfect Prediction')
plt.title('Prediction Scatter Plot', fontsize=12, fontweight='bold')
plt.xlabel('Actual Price ($)')
plt.ylabel('Predicted Price ($)')
plt.legend()
plt.grid(True, alpha=0.3)

plt.subplot(3, 2, 5)
errors = test_predictions_original - y_test_original
plt.hist(errors, bins=30, color='skyblue', edgecolor='black')
plt.axvline(x=0, color='r', linestyle='--', linewidth=2, label='Zero Error')
plt.title('Prediction Error Distribution', fontsize=12, fontweight='bold')
plt.xlabel('Prediction Error ($)')
plt.ylabel('Frequency')
plt.legend()
plt.grid(True, alpha=0.3)

plt.subplot(3, 2, 6)
recent_window = min(50, len(y_test_original))
plt.plot(range(recent_window), y_test_original[-recent_window:],
         label='Actual', marker='o', linewidth=2, markersize=4)
plt.plot(range(recent_window), test_predictions_original[-recent_window:],
         label='Predicted', marker='s', linewidth=2, markersize=4)
plt.title(f'Recent {recent_window} Predictions', fontsize=12, fontweight='bold')
plt.xlabel('Sample Index')
plt.ylabel('Stock Price ($)')
plt.legend()
plt.grid(True, alpha=0.3)

plt.tight_layout()
plt.savefig('stock_prediction_analysis.png', dpi=300, bbox_inches='tight')
print("Visualizations saved to 'stock_prediction_analysis.png'")

print("\n15. MODEL SUMMARY")
print("-"*80)
print(f"Total trainable parameters: {model.count_params():,}")
print(f"Input features: {X_train.shape[1]}")
print(f"Training samples: {X_train.shape[0]}")
print(f"Test samples: {X_test.shape[0]}")
print(f"Best validation loss: {min(history.history['val_loss']):.6f}")
print(f"Test RMSE: ${test_rmse:.4f}")
print(f"Trend prediction accuracy: {trend_accuracy:.2f}%")

print("\n" + "="*80)
print("STOCK MARKET PREDICTION COMPLETE")
print("="*80)

model.save('stock_prediction_model.h5')
print("\nModel saved as 'stock_prediction_model.h5'")

"""
================================================================================
COMPREHENSIVE EXPLANATION: STOCK MARKET PREDICTION USING ARTIFICIAL NEURAL NETWORK
================================================================================

1. PROJECT OVERVIEW:
   This program predicts future stock prices and trends using a deep Artificial
   Neural Network (ANN). It analyzes historical stock data including open, high,
   low, close prices, and volume to forecast next-day closing prices.

2. DATA PREPROCESSING:
   - Loads stock data with columns: date, open, high, low, close, volume, Name
   - Converts date strings to datetime format for proper time-series handling
   - Sorts data chronologically (critical for time-series predictions)
   - Removes any missing/null values to ensure data quality
   - Handles "########" entries which appear in the dataset as invalid dates

3. FEATURE ENGINEERING (17 Technical Indicators):

   a) Price Features:
      - price_change: Difference between closing and opening prices
      - price_change_pct: Percentage change during the day
      - high_low_diff: Daily price range (volatility indicator)
      - daily_return: Percentage change from previous close

   b) Moving Averages (Trend Indicators):
      - MA_5: 5-day moving average (short-term trend)
      - MA_10: 10-day moving average (medium-term trend)
      - MA_20: 20-day moving average (long-term trend)
      Used by traders to identify trend direction and support/resistance levels

   c) Volatility:
      - volatility: 5-day standard deviation of closing prices
      Measures price fluctuation intensity

   d) RSI (Relative Strength Index):
      - Momentum oscillator measuring overbought/oversold conditions
      - Values: 0-100 (>70 overbought, <30 oversold)
      - 14-day period standard

   e) Volume Analysis:
      - volume_change: Percentage change in trading volume
      High volume confirms trend strength

   f) Candlestick Patterns:
      - upper_shadow: Distance from high to body top
      - lower_shadow: Distance from body bottom to low
      Indicate buyer/seller pressure

   g) Target Variable:
      - target: Next day's closing price (shift(-1))
      This is what the model learns to predict

4. DATA NORMALIZATION:
   - Uses MinMaxScaler to scale all features to 0-1 range
   - Prevents features with larger values from dominating the learning
   - Essential for neural networks to converge efficiently
   - Separate scalers for features (X) and target (y) to allow inverse transformation

5. TRAIN-TEST SPLIT:
   - 80% training, 20% testing
   - shuffle=False maintains chronological order (crucial for time-series)
   - Tests model on most recent unseen data (realistic trading scenario)

6. ARTIFICIAL NEURAL NETWORK ARCHITECTURE:

   Layer 1 (Input): 128 neurons, ReLU activation
   - Takes all 17 features as input
   - ReLU (Rectified Linear Unit): f(x) = max(0, x)
   - Dropout 20%: Randomly disables neurons to prevent overfitting

   Layer 2 (Hidden): 64 neurons, ReLU activation
   - Learns complex feature combinations
   - Dropout 20%

   Layer 3 (Hidden): 32 neurons, ReLU activation
   - Further abstraction of patterns
   - Dropout 10%

   Layer 4 (Hidden): 16 neurons, ReLU activation
   - Final feature refinement

   Layer 5 (Output): 1 neuron, Linear activation
   - Produces continuous price prediction
   - Linear activation: f(x) = x (no transformation)

   Architecture reasoning:
   - Pyramidal structure (128→64→32→16→1) progressively reduces dimensions
   - Multiple layers learn hierarchical patterns (simple→complex)
   - Dropout prevents overfitting by forcing redundant learning paths

7. MODEL COMPILATION:
   - Optimizer: Adam (Adaptive Moment Estimation)
     * learning_rate=0.001: Controls step size during weight updates
     * Automatically adjusts learning rates per parameter
     * Combines momentum and RMSProp advantages

   - Loss Function: Mean Squared Error (MSE)
     * Measures average squared difference between predictions and actuals
     * Penalizes large errors more heavily
     * MSE = (1/n) * Σ(actual - predicted)²

   - Metrics: Mean Absolute Error (MAE)
     * Average absolute difference (easier to interpret than MSE)
     * MAE = (1/n) * Σ|actual - predicted|

8. TRAINING PROCESS:
   - Epochs: 200 maximum iterations through entire dataset
   - Batch Size: 32 samples processed before weight update
     * Smaller batches = more updates but noisier
     * 32 is good balance for this dataset size

   - Validation Split: 20% of training data for validation
     * Monitors performance on unseen data during training
     * Prevents overfitting

   - Early Stopping Callback:
     * Monitors validation loss
     * Stops training if no improvement for 20 epochs
     * Restores best weights (prevents overfitting)
     * Saves training time

9. EVALUATION METRICS:

   a) Mean Squared Error (MSE):
      - Average of squared prediction errors
      - Units: squared dollars ($²)
      - Sensitive to outliers

   b) Root Mean Squared Error (RMSE):
      - Square root of MSE
      - Same units as target variable ($)
      - More interpretable than MSE
      - Lower is better

   c) Mean Absolute Error (MAE):
      - Average absolute difference
      - In dollars ($)
      - Less sensitive to outliers than RMSE
      - Directly interpretable

   d) R² Score (Coefficient of Determination):
      - Proportion of variance explained by model
      - Range: -∞ to 1 (1 = perfect, 0 = baseline, <0 = worse than mean)
      - R² = 1 - (SS_residual / SS_total)

   e) MAPE (Mean Absolute Percentage Error):
      - Average percentage error
      - Easy to understand (e.g., 2.5% average error)
      - MAPE = (100/n) * Σ|(actual - predicted) / actual|

10. TREND PREDICTION ACCURACY:
    - Measures ability to predict price direction (up/down/neutral)
    - More important for trading than exact price prediction
    - Calculated by comparing sign of actual vs predicted changes
    - Trading strategy can be built on trend accuracy alone

11. PREDICTION FUNCTION (predict_next_day):
    - Takes latest feature values as input
    - Scales input using fitted scaler
    - Runs through trained neural network
    - Inverse transforms output to original price scale
    - Returns next day's predicted closing price

12. TRADING RECOMMENDATIONS:
    Based on predicted price change:
    - Positive change → UPWARD trend → Consider BUYING
    - Negative change → DOWNWARD trend → Consider SELLING/HOLDING
    - Near zero → NEUTRAL → HOLD position

    Note: These are basic recommendations. Real trading requires:
    - Risk management strategies
    - Multiple confirmation indicators
    - Market sentiment analysis
    - Fundamental analysis

13. VISUALIZATIONS CREATED:

    a) Training Loss Curves:
       - Shows model learning progress
       - Training vs validation loss comparison
       - Identifies overfitting (diverging curves)

    b) MAE Curves:
       - Tracks prediction accuracy improvement
       - Monitors validation performance

    c) Actual vs Predicted Prices:
       - Line plot comparing predictions to reality
       - Visual assessment of model accuracy

    d) Scatter Plot:
       - Each point: (actual, predicted)
       - Perfect line: y=x (45-degree line)
       - Points close to line = good predictions

    e) Error Distribution:
       - Histogram of prediction errors
       - Should be centered at zero
       - Normal distribution indicates unbiased predictions

    f) Recent Predictions:
       - Zooms into last 50 predictions
       - Shows model performance on most recent data

14. NEURAL NETWORK ADVANTAGES FOR STOCK PREDICTION:

    a) Non-linear Relationships:
       - Markets have complex, non-linear patterns
       - ANNs can model these without explicit programming

    b) Feature Interactions:
       - Automatically learns how features interact
       - Hidden layers discover combined effects

    c) Adaptability:
       - Can be retrained with new data
       - Learns evolving market patterns

    d) Multiple Input Handling:
       - Processes 17+ features simultaneously
       - Considers all technical indicators together

15. LIMITATIONS AND CONSIDERATIONS:

    a) Market Unpredictability:
       - Stock prices influenced by news, events, sentiment
       - Historical patterns don't guarantee future results
       - Black swan events unpredictable

    b) Overfitting Risk:
       - Model might memorize training data
       - Dropout and early stopping help mitigate
       - Regular retraining required

    c) Data Quality:
       - Predictions only as good as input data
       - Missing data, errors affect accuracy
       - Need clean, reliable data sources

    d) Not a Complete Trading System:
       - Should be combined with:
         * Fundamental analysis
         * Risk management
         * Portfolio diversification
         * Stop-loss orders

16. TECHNICAL DETAILS:

    - Language: Python
    - Framework: TensorFlow/Keras (industry-standard deep learning)
    - Data Processing: Pandas, NumPy
    - Visualization: Matplotlib, Seaborn
    - Scaling: Scikit-learn MinMaxScaler

17. MODEL PERSISTENCE:
    - Saves trained model as 'stock_prediction_model.h5'
    - Can be loaded later without retraining
    - Format: HDF5 (Hierarchical Data Format)
    - Usage: model = load_model('stock_prediction_model.h5')

18. REAL-WORLD APPLICATIONS:

    a) Algorithmic Trading:
       - Automated buy/sell decisions
       - High-frequency trading systems

    b) Portfolio Management:
       - Asset allocation optimization
       - Risk assessment

    c) Risk Management:
       - Volatility forecasting
       - Value at Risk (VaR) calculation

    d) Investment Research:
       - Trend identification
       - Entry/exit point suggestions

19. PERFORMANCE INTERPRETATION:

    Good Model Indicators:
    - RMSE < 5% of average stock price
    - R² Score > 0.8
    - Trend accuracy > 60%
    - MAPE < 5%
    - Training and validation loss converging

    Warning Signs:
    - Large gap between train and test metrics (overfitting)
    - Trend accuracy < 50% (worse than random)
    - Increasing validation loss while training loss decreases

20. FUTURE ENHANCEMENTS:

    a) LSTM Networks:
       - Better for sequential time-series data
       - Captures long-term dependencies
       - Remembers important historical patterns

    b) Additional Features:
       - Sentiment analysis from news/social media
       - Macroeconomic indicators (GDP, inflation)
       - Company fundamentals (P/E ratio, earnings)
       - Market indices correlation

    c) Ensemble Methods:
       - Combine multiple models
       - Random Forest + ANN + LSTM
       - Improves robustness

    d) Real-time Prediction:
       - Integration with live data feeds
       - Automatic retraining pipeline
       - API for trading platforms

    e) Multi-step Forecasting:
       - Predict multiple days ahead
       - Predict high/low in addition to close
       - Confidence intervals for predictions

================================================================================
DISCLAIMER: This model is for educational purposes. Stock market predictions
are inherently uncertain. Always consult financial advisors before making
investment decisions. Past performance does not guarantee future results.
================================================================================
"""
```
